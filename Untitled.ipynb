```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
from sklearn.metrics import classification_report
```

```python
cols = ["flength", "fwidth","fsize", "fconc", "fconc1", "fasym", "fm3long", "fm3trans", "falpha", "fdist", "class"]
df = pd.read_csv("magic04.data", encoding="iso-8859-1", names= cols)
df.head()
```

```python
df["class"] = (df["class"] == "g").astype(int)
```

```python

df.head()

```

```python
for label in cols[:-1]:
    plt.hist(df[df["class"] == 1][label], color = 'red', label = 'gamma', alpha = 0.7, density = True)
    plt.hist(df[df["class"] == 0][label], color = 'blue', label = 'hadron', alpha = 0.7, density = True)
    plt.title(label)
    plt.ylabel("probablity")
    plt.xlabel(label)
    plt.legend()
    plt.show()
```

```python
train, valid, test = np.split(df.sample(frac =1), [int(0.6*len(df)), int(0.8 * len(df))])
```

```python
def scale_dataset(dataframe, oversample = False):
    # Extract features (X) and target (y) columns correctly
    x = dataframe[dataframe.columns[:-1]].values
    y = dataframe[dataframe.columns[-1]].values

    scaler = StandardScaler()
    x = scaler.fit_transform(x)

    if oversample:
        ros = RandomOverSampler()
        x ,y = ros.fit_resample(x, y)
        
    data = np.hstack((x, np.reshape(y, (-1, 1))))
    
    return data, x, y
```

```python
train, x_train , y_train =  scale_dataset(train, oversample=True)
valid, x_valid , y_valid =  scale_dataset(valid, oversample=False)
test, x_test , y_test =  scale_dataset(test, oversample=False)
```

```python
#KNN implementation 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(x_train, y_train)
 
```

```python
y_pred = knn_model.predict(x_test)
y_pred
```

```python
y_test
```

```python
print(classification_report(y_test, y_pred))
```
