```python
import pandas as pd

import numpy as np
```

```python
df = pd.read_csv("Book1.csv", encoding="UTF-8")
df.head(10)
```

```python
df.columns = df.columns.str.strip()
df.head()
```

```python
df['Age'] = df['Age/Sex'].str.extract('(\d+)').astype(float)
# df['Sex'] = df['Age/Sex'].str.extract('([mf])', flags=re.IGNORECASE).str.upper()
df['Weight'] = df['Weight{ in kg}'].str.extract('(\d+)').astype(float)
df['Height_feet'] = df['Height'].str.extract("(\d+)'").astype(float)
df['Height_inches'] = df['Height'].str.extract("'(\d+)").fillna(0).astype(float)
df['Height_cm'] = (df['Height_feet'] * 30.48) + (df['Height_inches'] * 2.54)
```

```python
df.head()
```

```python
df['BMI'] = df['Weight'] / ((df['Height_cm']/100) ** 2)
```
